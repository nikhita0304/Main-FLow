```python
df.HeartDisease.value_counts().plot(kind = 'bar' , color=["orchid", "salmon"])
plt.title("Heart Disease Values")
plt.xlabel("1 = Heart Disease, 0 = No Heart Disease")
plt.ylabel("Count")
plt.show()
```
